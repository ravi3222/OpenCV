```python
import cv2
import numpy as np

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

if face_cascade.empty():
    raise IOError('Unable to load the face cascade classifier xml file')
if eye_cascade.empty():
    raise IOError('Unable to load the eye cascade classfier xml file')

cap = cv2.VideoCapture(0)
sunglass_img = cv2.imread('sg.png')
count=0
while True:
    ret,frame = cap.read()
    frame = cv2.resize(frame,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
    vh,vw = frame.shape[:2]
    vh,vw = int(vh),int(vw)
    
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3,minNeighbors = 1)
    
    centers = []
    
    for (x,y,w,h) in faces:
        print(x,y,w,h)
        roi_gray = gray[y:y+h,x:x+w]
        roi_color = faces[y:y+h,x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (x_eye,y_eye,w_eye,h_eye) in eyes:
            centers.append((x+ int(x_eye+0.5*w_eye),y+ int(y_eye+0.5*h_eye)))
            print('center :'+str(centers))
        print('-----------')
    
    if len(centers)>1:
        count=count+1
        print('Eyes detected')
        h,w = sunglass_img.shape[:2]
        eye_distance = abs(centers[1][0] - centers[0][0])
        sunglass_width = 2.12 * eye_distance
        scalingFactor = sunglass_width/w
        print(scalingFactor,eye_distance)
        overlay_sunglass = cv2.resize(sunglass_img,None,fx=scalingFactor,fy=scalingFactor,interpolation=cv2.INTER_AREA)
        x = centers[0][0] if centers[0][0] < centers[1][0] else centers[1][0]
        x -= int(0.26*overlay_sunglass.shape[1])
        y += int(0.26*overlay_sunglass.shape[0])
        h,w = overlay_sunglass.shape[:2]
        h,w = int(h),int(w)
        frame_roi = frame[y:y+h,x:x+w]
        gray_overlay_sunglass = cv2.cvtColor(overlay_sunglass,cv2.COLOR_BGR2GRAY)
        ret,mask = cv2.threshold(gray_overlay_sunglass,180,255,cv2.THRESH_BINARY_INV)
        mask_inv = cv2.bitwise_not(mask)
        try:
            masked_face = cv2.bitwise_and(overlay_sunglass,overlay_sunglass,mask=mask)
            masked_frame = cv2.bitwise_and(frame_roi, frame_roi,mask=mask_inv)
        except cv2.error as e:
            print('Ignoring arithmentic exceptions: '+ str(e))
        #raise e
        # add the two images to get the final output
        frame[y:y+h, x:x+w] = cv2.add(masked_face, masked_frame)
    else:
        print('Eyes not detected')
    print(count)
    cv2.imshow('Eye Detector', frame)
    c = cv2.waitKey(1)
    if c == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
